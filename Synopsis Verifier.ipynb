```python
from SentenceParserPython3 import SentenceParser
import pandas as pd
```

```python
SP = SentenceParser()
SP.readfile('./nvbugs_Synopsis.csv','csv', header =0)
SP.data = SP.data.drop('Unnamed: 0', 1)
```

```python
def overallreplacement(df, column, new_name):
    df[new_name] = df[column].str.replace(r'[\n\r\t]+', ' ')
    df[new_name] = df[column].str.replace(r'[\[\]/:?\\"|]+', ' ')
```

```python
overallreplacement(SP.data, 'Synopsis','Cleaned')
SP.data
```
